# 리스트 컴프리핸션

## Exercises (1)
* 아래 리스트에서 음수와 0만 필터링하여 새 리스트를 만드세요.

```python
numbers = [-4, -3, -2, -1, 0, 2, 4, 6]
```

```python
negatives_and_zero = [i for i in numbers if i <= 0]
print(negatives_and_zero)
# [-4, -3, -2, -1, 0]
```

## Exercises (2)
* 중첩 리스트 평탄화 (Flattening a 3D List): 아래와 같은 3중 리스트를 1차원 리스트로 변환하세요.

```python
list_of_lists = [[[1, 2, 3]], [[4, 5, 6]], [[7, 8, 9]]]
```

```python
flattened = [num for sublist1 in list_of_lists for sublist2 in sublist1 for num in sublist2]
print(flattened)
# [1, 2, 3, 4, 5, 6, 7, 8, 9]
```
