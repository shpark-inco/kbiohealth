# 02 자료형

파이썬에는 여러 가지 데이터 타입이 있습니다. 데이터 타입을 확인할 때는 type() 내장 함수를 사용한다.

## 02-1 데이터 타입 확인
특정 데이터나 변수의 데이터 타입을 확인할 때는 type() 함수를 사용합니다.

```python
first_name = 'Asabeneh'     # 문자열(str)
last_name = 'Yetayeh'       # 문자열(str)
country = 'Finland'         # 문자열(str)
city = 'Helsinki'           # 문자열(str)
age = 250                   # 정수(int)

# 데이터 타입 출력
print(type('Asabeneh'))           # <class 'str'>
print(type(first_name))           # <class 'str'>

print(type(10))                   # <class 'int'>
print(type(3.14))                 # <class 'float'>
print(type(1 + 1j))               # <class 'complex'>

print(type(True))                 # <class 'bool'>

print(type([1, 2, 3, 4]))         # <class 'list'>
print(type({'name':'Asabeneh'}))  # <class 'dict'>
print(type((1,2)))                # <class 'tuple'>
print(type(zip([1,2],[3,4])))     # <class 'zip'>

# [Shift]+[Enter]로 셸을 실행해보아요.
```

형 변환(casting)이란, 한 데이터 타입을 다른 데이터 타입으로 변환하는 것입니다.

* 사용하는 함수: int(), float(), str(), list(), set() 등
* 예를 들어, 문자열로 된 숫자는 연산을 하기 전에 정수(int)나 실수(float) 로 변환해야 한다.
* 숫자와 문자열을 합치고 싶다면, 먼저 숫자를 문자열(str) 로 변환해야 한다.

```python
# int -> float
num_int = 10
print('num_int:', num_int)           # 10
num_float = float(num_int)
print('num_float:', num_float)       # 10.0

# float -> int
gravity = 9.81
print(int(gravity))                  # 9

# int -> str
num_int = 10
num_str = str(num_int)
print(num_str)                       # '10'

# str -> float, int
num_str = '10.6'
num_float = float(num_str)
print('num_float:', num_float)       # 10.6
num_int = int(num_float)
print('num_int:', num_int)           # 10

# str -> list
first_name = 'Asabeneh'
first_name_to_list = list(first_name)
print(first_name_to_list)            # ['A', 's', 'a', 'b', 'e', 'n', 'e', 'h']
```

## 02-2 연산 (기본형, 배열형)

|Operator|Name|Example|
|:--:|:--:|:--:|
|+|Addition|x + y|
|-|Subtraction|x - y|
|*|Multiplication|x * y|
|/|Division|x / y|
|%|나머지 연산자(Modulus)|x % y|
|**|거듭제곱(Exponentiation)|x ** y|
|//|몫 연산자(Floor division)|x //y |

```python
# 숫자형

## 정수
print('Addition: ', 1 + 2)        # 3
print('Subtraction: ', 2 - 1)     # 1
print('Multiplication: ', 2 * 3)  # 6
print('Division: ', 4 / 2)        # 2.0  -> 파이썬에서는 나누기 연산이 항상 실수(float)를 반환
print('Division: ', 6 / 2)        # 3.0
print('Division: ', 7 / 2)        # 3.5
print('Division without the remainder: ', 7 // 2)   # 3 -> 몫만 반환
print('Division without the remainder: ', 7 // 3)   # 2
print('Modulus: ', 3 % 2)         # 1 -> 나머지
print('Exponentiation: ', 2 ** 3) # 8 -> 2 * 2 * 2

## 실수
print('Floating Point Number, PI', 3.14)
print('Floating Point Number, gravity', 9.81)

## 복소수
print('Complex number: ', 1 + 1j)
print('Multiplying complex numbers: ', (1 + 1j) * (1 - 1j))
```

Question: 리스트와 튜플은 무엇이 다른가요?

* 리스트와 튜플은 서로 다른 자료형을 담을 수 있는 **순서가 있는** 컬렉션이다.
* 튜플은 **불변(immutable)** 이므로, 한 번 생성하면 값을 변경할 수 없다.
* 따라서 add, insert, remove 같은 메서드는 사용할 수 없다.

-> 튜플은 불변이므로 수정하려면 리스트로 변환 후 수정 가능 <br>
-> 튜플의 개별 항목은 삭제 불가. 전체는 삭제 가능
